```python
products=pd.read_csv("Products.csv")
products.isnull().sum()
```

```python
products.dtypes
```

```python
products['Unit Price USD'] = products['Unit Price USD'].replace('[\$,]', '', regex=True).astype(float)
```

```python
products['Unit Cost USD'] = products['Unit Cost USD'].replace('[\$,]', '', regex=True).astype(float)
```

```python
products.isnull().sum()
```

```python
products['Profit per Unit'] = products['Unit Price USD'] - products['Unit Cost USD']
```

```python
products['profit_per_Unit'] = products['profit_per_Unit'].apply(lambda x: f"{x:.2f}")
```

```python
products
```

```python
products.describe()
```

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

host = "localhost"
port = "3306"
username = "root"
password = "****"
database = "capstone_2"

engine_string = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_string)

csv_files = ['sales.csv', 'customers.csv', 'products.csv', 'stores.csv','exchange.csv']  
table_names = ['sales', 'customers', 'products', 'stores','exchange'] 


for csv_file, table_name in zip(csv_files, table_names):
    try:
        df = pd.read_csv(csv_file)
        df.to_sql(table_name, engine, index=False, if_exists='replace') 
        print(f"Successfully uploaded {csv_file} to table {table_name}")
    except Exception as e:
        print(f"Error uploading {csv_file}: {e}")

print("All files processed")
```
